```python
"""
User will enter a floating point number. We need to find the integer portion before the decimal, and check if that is even or odd.
"""
x= float(input("Enter a real number: "))
y= round(x)
if x>0:
    if y>x:
        intPortion= y-1
    else:
        intPortion= y
else:
    if y<x:
        intPortion= y+1
    else:
        intPortion= y
if intPortion%2==0:
    print("Even Number.")
else:
    print("Odd Number.")
```

```python
"""
Given a list i.e, L=[2,3,4,7,6,-4,-1,9], arrange the list in sorted order and print the sorted list.
"""
L=[2,3,4,7,6,-4,-1,9]
for j in range (len(L)):
    m=L[j]
    idx=j
    c=j
    for i in range (j, len(L)):
        if L[i]<m:
            m=L[i]
            idx=c
        c+=1
    tmp=L[j]
    L[j]=m
    L[idx]=tmp
print(L)
```
